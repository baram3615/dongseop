```python
import pandas as pd
df = pd.read_csv('d:/bitmex_data_1m.csv', delimiter=',')
```

```python
import datetime 
import time

#df의 timestamp 행을 Windowing 하기위해 int 형태로 바꾸는 함수
def convert_to_timestamp(row):
    return time.mktime(datetime.datetime.strptime(row, '%Y-%m-%d %H:%M:%S').timetuple())

#+00:00 를 없애고 KST 로 바꾸는 함수
def convert_gmt_to_kst(row):
    return str(datetime.datetime.strptime(row, '%Y-%m-%d %H:%M:%S+00:00')+datetime.timedelta(hours=9)-datetime.timedelta(minutes=1))

def convert_timestamp_to_datetime_str(row):
    return str(datetime.datetime.fromtimestamp(row))
#time.mktime(datetime.datetime.strptime(df['timestamp'][0], '%Y-%m-%d %H:%M:%S+00:00').timetuple())
```

```python
#2023년도 데이터만 갖고 놀자
df_2023 = df.loc[df.timestamp >= '2022-12-31 15:01:00']  #일단 GMT 니까 1분 뒤로 조정할 걸 생각하고 +1분부터 가져오면 된다.

#kst 로 바꾸고 기존 gmt 형태는 drop
df_2023['timestamp_kst'] = df_2023['timestamp'].apply(convert_gmt_to_kst)

df_2023 = df_2023[['timestamp_kst','low','high','close']]

df_2023 = df_2023.reset_index()

#int 형태의 timestamp 열도 추가
df_2023['timestamp_int'] = df_2023['timestamp_kst'].apply(convert_to_timestamp)

#다시 필요한 컬럼만 정돈
df_2023 = df_2023[['timestamp_kst','timestamp_int','low','high','close']]
```

```python
#15분봉의 timestamp 값이 06:30:00 이라면  06:30:00 부터 06:44:59 까지
#즉 2023-01-01 00:00:00+00:00 일 경우, 과거의 15개를 쓰는것이 아니고 나를 포함한 미래의 15개를 써야함

# 일반적인 window15 rolling의 결과는 
#2023-01-01 00:14:00+00:00 → 2023-01-01 00:00:00+00:00 ~ 2023-01-01 00:14:00+00:00 
#2023-01-01 00:15:00+00:00 → 2023-01-01 00:01:00+00:00 ~ 2023-01-01 00:15:00+00:00 

#결론적으로 
#timestamp 중에 필요한 건 900이란 값을 %를 했을 때 0이 되는 값 2023-01-01 00:00:00+00:00
#low 는 window 15 를 min() 한 것을 14번 뒤로 shift 했을 때 나오는 값, 
#high도 마찬가지고 window 15를 max() 한 것을 14번 뒤로 shift 했을 때 나오는 값,
#close 는 timestamp : 2023-01-01 00:00:00+00:00 일 때, > 1분봉에서의 2023-01-01 00:14:00+00:00 의 최종 close 를 그대로 쓰면 되기 때문에
#rolling 하지 않고 그냥 14개 뒤로 shift 한 값의 close 를 사용하면 됨..

#함수로 변환할 때 필요변수 : 
#1) timestamp_base_gap 900 (15분 간격의 값이라는것을 표시)
#2) shift = 15

#df_2023['timestamp','low','high','15m_low','15m_high']
```

```python
#최종 15분봉의 형태구현
final_df = df_2023[['low']].rolling(window=15).min()     #15개를 min값하면 모수가 되지 못한 맨 앞의 14개는 NaN이된다.
final_df['high'] = df_2023[['high']].rolling(window=15).max() #15개를 max값해도 동일하게 모수가 되지 못한 맨 앞의 14개는 NaN이된다.
final_df['close'] = df_2023['close']                         #어짜피 맨 앞의 14개는 지워줄 거고, 맨 앞의 값이 되는 2023-01-01 00:14:00
                                                             #가 아래 timestamp_int 로는 2023-01-01 00:00:00 라는 새로운 값이 될 것이다.
final_df['timestamp_int'] = df_2023[['timestamp_int']]-840  #14칸(840/60)을 위로 올려준다음 맨 앞의 14개를 지워버리면 일단 clase 빼고는 완성된다.

final_df = final_df[14:] #window -1 값만큼 버리고
final_df['timestamp_kst'] = final_df['timestamp_int'].apply(convert_timestamp_to_datetime_str)
final_df = final_df.reset_index()[['timestamp_kst','low','high','close','timestamp_int']]#리셋
```
